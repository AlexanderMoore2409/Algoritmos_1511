```python
import time
def sumar( n ):
    suma = 0
    if n>0:
        for x in range(n+1):
            suma += x

    print( suma )

print(time.time())
ini = time.time()
sumar(10000000)
fin = time.time()
print(f"Tiempo total:{(fin - ini)} milisegundos")

```

```python
import time
def sumar2( n ):
    datos = [x for x in range(n+1)]
    #print(datos)
    suma = 0
    if n>0:
        for x in range(n+1):
            suma += datos[x]

    print( suma )

print(time.time())
ini = time.time()
sumar2(10000000)
fin = time.time()
print(f"Tiempo total:{(fin - ini)} milisegundos")
```

```python
import sys
import time
def sumar3( n ):
    if len(n)==1:
        return n[0]
    else:
        return sumar3(n[1::]) + n[0]

sys.setrecursionlimit(10000)
print("inicio")
ini = time.time()
n=100000
datos = [x for x in range(n+1)]
#print(datos)
res = sumar3(datos)
print(res)
fin = time.time()
print(f"Tiempo total:{(fin - ini)} milisegundos")
```
